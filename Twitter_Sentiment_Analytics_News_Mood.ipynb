```python
# Array to hold sentiment
sentiment_array = []
sentiment_array_ALL = []
Media_Compounds_Mean = []

# Variable for holding the oldest tweet
oldest_tweet = ""
counter = 1

# Loop through all target users
for target in target_terms:
    print(target)
    
    # Run search around each tweet
    public_tweets = api.search(target, count=100, result_type="recent")
    
    # Loop through all tweets
    for tweet in public_tweets["statuses"]:
        
        
        #print (json.dumps(tweet, indent = 4))
        # Run Vader Analysis on each tweet
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        
        
        # Add sentiments for each tweet into an array
        sentiment_array.append({"Name": target,
                                "Compound": compound,
                                "Positive": pos,
                                "Negative": neu,
                                "Neutral": neg,
                                "Tweet Text":tweet["text"],
                                "Tweets Ago": counter})
        
        
        sentiment_array_ALL.append({"Name": target,
                                    "Compound": compound,
                                    "Positive": pos,
                                    "Negative": neu,
                                    "Neutral": neg,
                                    "Tweet Text":tweet["text"],
                                    "Tweets Ago": counter})
        
        counter = counter + 1         
        
    sentiments_pd = pd.DataFrame.from_dict(sentiment_array)
    sentiments_pd.to_csv(target + "sentiments.csv")
    
    sentiment_array = []
    
    
    #Create plot for each News Agency
    plt.plot(np.arange(len(sentiments_pd["Compound"])),
             sentiments_pd["Compound"], marker="o", linewidth=0.5,
             alpha=0.8, label = "target")

   # Incorporate the other graph properties
    plt.title("Sentiment Analysis of Tweets (%s) for %s" % (time.strftime("%x"), target))
    plt.ylabel("Tweet Polarity")
    plt.xlabel("Tweets Ago")
    plt.show()          
    plt.savefig(target+".png")
   
   
sentiments_pd_ALL = pd.DataFrame.from_dict(sentiment_array_ALL)      
sentiments_pd_ALL.head(500)
```

Now that we have a DataFrame for all of the sentiments,

```python
# Each target_user, define compound_polarity
# target_terms = ["@BBC", "@CBS", "@CNN", "@Fox", "@nytimes"]

#@BBC:
BBC_score = sentiments_pd_ALL[sentiments_pd_ALL["Name"] == "@BBC"]
#BBC_score.head()

#@CBS:
CBS_score = sentiments_pd_ALL[sentiments_pd_ALL["Name"] == "@CBS"]
#CBS_score.head()

#@CNN:
CNN_score = sentiments_pd_ALL[sentiments_pd_ALL["Name"] == "@CNN"]
#CNN_score.head()

#@Fox:
Fox_score = sentiments_pd_ALL[sentiments_pd_ALL["Name"] == "@Fox"]
#Fox_score.head()

#@nytimes
nytimes_score = sentiments_pd_ALL[sentiments_pd_ALL["Name"] == "@nytimes"]
nytimes_score.head(2)
#nytimes_score.columns
```

```python
# Build the scatter plots for each target_user: Tweets Ago = x-axis, Compound = y-axis, 
# alpha = transparency of bubble. This will apply to all target_user

plt.scatter(x=BBC_score["Tweets Ago"],
            y=BBC_score["Compound"],
            facecolor="coral", edgecolor="black",
            linewidths=1, marker="o",
            alpha =0.8, label="BBC")
plt.scatter(x=CBS_score["Tweets Ago"],
            y=CBS_score["Compound"],
            facecolor="skyblue", edgecolor="black",
            linewidths=1, marker="o",
            alpha =0.8, label="CBS")
plt.scatter(x=CNN_score["Tweets Ago"],
            y=CNN_score["Compound"],
            facecolor="yellow", edgecolor="black",
            linewidths=1, marker="o",
            alpha =0.8, label="CNN")
plt.scatter(x=Fox_score["Tweets Ago"],
            y=Fox_score["Compound"],
            facecolor="green", edgecolor="black",
            linewidths=1, marker="o",
            alpha =0.8, label="Fox")
plt.scatter(x=nytimes_score["Tweets Ago"],
            y=nytimes_score["Compound"],
            facecolor="blue", edgecolor="black",
            linewidths=1, marker="o",
            alpha =0.8, label="nytimes")


# Scatter Properties 
plt.title("Sentiment Analysis of Tweets ("+ date +")")
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")


lgn = plt.legend(fontsize="small", mode="Expanded",
                numpoints=1, scatterpoints=1, loc='upper left', bbox_to_anchor=(1,1), 
                title ="News Source",
                labelspacing =0.5) 

plt.show()
plt.savefig('Sentiment_scatter_plot.png')
```

Now we must plot a bar graph that displays the average compound score for each target_user. we use .groupby Name .mean() ["Compound"].

```python
# Make a list of Average Compound Score of each target_user
Avg_Compound = round(sentiments_pd_ALL.groupby(["Name"]).mean()["Compound"],2)
Avg_Compound_df = pd.DataFrame(Avg_Compound)
Avg_Compound_df.head()
```

```python
x_axis = np.arange(len(Avg_Compound))
#xlabels = Avg_Compound.index
xlabels = ["BBC", "CBS", "CNN", "Fox", "nytimes"]

count = 0

for Compound in Avg_Compound:
    plt.text(count, Compound+0.05, str(round(Compound,2)))
    count = count + 1
    

plt.bar(x_axis, Avg_Compound, tick_label = xlabels,
        color=["coral", "skyblue", "yellow", "green", "blue"], align="edge")

# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value+0.4 for value in x_axis]
plt.xticks(tick_locations, xlabels)

# Sets the x limits of the current chart
plt.xlim(-0.25, len(x_axis))

# Sets the y limits of the current chart
plt.ylim(-.025, 0.17)


plt.title('Overall Media Sentiment based on Twitter ('+ date +")")
plt.xlabel("News Organization")
plt.ylabel("Tweet Polarity")
plt.savefig('Sentiment_barchart.png')
plt.show()
```
